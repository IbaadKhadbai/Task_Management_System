```python
tasks = []

def menu():
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")

menu()
```

```python
def add_task():
    task = input("Enter task: ")
    tasks.append(task)
    print("Task added")
```

```python
def view_tasks():
    for task in tasks:
        print(task)
```

```python
def complete_task():
    task = input("Enter task to complete: ")
    if task in tasks:
        tasks.remove(task)
        print("Task completed")
```

```python
while True:
    menu()
    choice = input("Choose option: ")

    if choice == "1":
        add_task()

    elif choice == "2":
        view_tasks()

    elif choice == "3":
        complete_task()
```
